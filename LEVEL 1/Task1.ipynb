```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
url="https://drive.google.com/file/d/1CCwMErhA8y7_IDBl7MlqfCPUt3hb87CP/view?usp=drive_link"
url="https://drive.google.com/uc?id=" + url.split("/")[-2]
df = pd.read_csv(url)
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df['Cuisines'])
```

```python
import pandas as pd
cuisine_count = df['Cuisines'].str.split(', ').explode().value_counts()
print(cuisine_count)
top_cuisine = cuisine_count.head(3)
print("The Top 3 Cuisines are:", top_cuisine)
```

```python
colours = ['green', 'orange', 'red']
plt.bar(top_cuisine.index, top_cuisine.values, color=colours)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top Three Cuisines')
plt.show()
```

```python
total_restaurant = len(df)
print("Total Restaurants:", total_restaurant)
top_cuisine10 = cuisine_count.head(10)
top_cuisine5 = cuisine_count.head(5)
percentages = (top_cuisine5 / total_restaurant) * 100
print("The Market Share of Top 5 Cuisines is:\n", percentages)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_cuisine5.index, percentages.values, color=colours)
plt.xlabel('Cuisine Name')
plt.ylabel('Percentage')
plt.title('Top Five Cuisines')
plt.show()
```
